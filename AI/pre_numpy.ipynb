```python
import numpy as np
ndarray = np.arange(12).reshape(3, 4)
print(ndarray)
print(ndarray.ndim)     #维度
print(ndarray.shape)    #行列数
print(ndarray.size)     #元素个数
print(ndarray.dtype)    #元素类型
print(ndarray.itemsize) #每个元素占用的字节数
print(ndarray.data)     #元素所在内存地址
```

```python
a = np.array((1,2,3))   #from tuple
b = np.array([1,2,3])   #from list
c = np.array([(1,2,3), (4,5,6)])   #from list of list
print(type(a))          #<type 'numpy.ndarray'>
print(type(b))    
print(c)      
```

```python
zero = np.zeros((3,4))   #全0数组
one = np.ones((3,4))     #全1数组
empty = np.empty((3,4))  #未初始化的数组，元素随机
print(zero)
print(one)
print(empty)
zero_int = np.zeros((3,4), dtype=np.int16)   #指定元素类型,默认为float64
print(zero_int)
```

```python
np.arange(10, 30, 5)   #等差数列，起始值、终止值、步长
np.arange(0, 2, 0.3)   #接收浮点数参数，但由于精度有限，结果不可预测
np.linspace(0, 2, 9)   #通常指定元素个数而不是给定步长
from numpy import pi
x = np.linspace(0, 2*pi, 100)   #在0~2*pi范围内返回100个均匀间隔的数
f = np.sin(x)   #计算sin(x)
import matplotlib.pyplot as plt
plt.plot(x, f)
plt.show()
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)   #元素级加法
print(a < 3)   #元素级比较
print(a * b)   # * 依然表示元素级乘法
print(a.dot(b))   #点积使用dot或者@
print(np.dot(a, b))
print(a @ b)
```

```python
b = np.arange(12).reshape(3,4)
print(b)
print(b.sum(axis=0))   #对列求和
print(b.sum(axis=1))   #对行求和
print(b.cumsum(axis=0))   #对列求累积和
print(b.cumsum(axis=1))   #对行求累积和
```

```python
a = np.arange(12)
print(a)
a[:6:2] = 13
print(a)
b = np.fromfunction(lambda x, y: y, (3, 4))   #根据函数创建数组
print(b)
```

```python
a = np.array([[[11,22,33],
               [12,13,14]],
              [[21,32,43],
               [15,16,17]]])
a.shape #(2,2,3) 表示2个2行3列的array
a[:,:,2]
a[...,2]    #...占位
for item in a:  #迭代默认针对第一维度
    print(item)

for each in a.flat: #.flat可以遍历所有元素
    print(each)
```
