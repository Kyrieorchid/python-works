# about copy and view
* 可变对象的<u>简单赋值</u>以及<u>函数传参</u>都是传递的<u>引用</u>，不会执行copy
* Different array objects can share the same data. The view method creates a new array object that looks at the same data.
* view方法、切片，都返回一个view
* copy方法才会完全复制一份，即deep copy

```python
a = np.array([1,2,3,4])
b = a.view()
print(id(a))
print(id(b))
b[0] = 2
print(a)    #different objects, but share the same data

s = a[:2]
s[:] = 10   #slicing operation actually returns a view
print(s)
print(a)

'''deep copy is often used after slicing,
for example, get final result from a huge intermediate result'''
intermediate_result = np.arange(10000)
final_result = intermediate_result[:100].copy()
del intermediate_result
```

# tricks of indexing

```python
# 用array作为indexes
a = np.arange(12) + 2
i = np.array([1,3,5])
print(a)
print(a[i])    #elements of a at the position i
j = np.array([[3,4],[9,7]]) #多维数组的话在一维基础上reshape
print(a[j])
```

```python
#多维index多维数组
palette = np.array([[0,0,0],            #black
                    [255,0,0],          #red
                    [0,255,0],          #green
                    [0,0,255],          #blue
                    [255,255,255]])     #white
image = np.array([[0,1,2,0],
                  [0,3,4,0]])
palette[image]

'''二维image，axis 1有两项，每一项都要从palette中的axis 1寻址，
    可以理解为用被寻址的array对应的内容去填充寻址数组的index，
    所以格式与寻址数组相同'''

#寻址array分开写
a = np.arange(12).reshape(3,4)
print(a)
i = np.array([[0,1], [1,2]]) #第一维寻址
print(a[i,:])
j = np.array([[2,1], [3,3]]) #第二维寻址
print(a[:,j])
print(a[i,j])
#还有很多......
```
